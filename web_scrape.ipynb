```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import time
import pandas as pd
```

```python
# Define browser and action setup
PATH = "f:\webdriver\chromedriver.exe"
driver = webdriver.Chrome(PATH)
```

```python
# URL
linkedin_url = "https://www.linkedin.com/jobs/search?keywords=Data%20Analyst&location=Indonesia&geoId=102478259&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0"
```

```python
# Action Steps
driver.maximize_window()
driver.get(linkedin_url)
```

```python
# Determine how many jobs we want to scrape and calculate how many time we need to scroll down
number_of_jobs = 300 #define how many jobs that we want to scrape
number_of_scroll = int(number_of_jobs/25)+1
print(number_of_scroll)
i = 1
driver.execute_script("return document.body.scrollHeight") #scroll to top
while i <= number_of_scroll:
    driver.execute_script("return document.body.scrollHeight")
    time.sleep(10) #waiting time
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #scroll to the bottom of page
    time.sleep(10)
    i = i + 1
    try:
        button = driver.find_element(By.XPATH,"/html/body/div[1]/div/main/section[2]/button").click()
        time.sleep(10)
        print("load more click")
    except:
        driver.execute_script("return document.body.scrollHeight")
        time.sleep(10)

print ("total jobs :")
jobs = driver.find_element(By.CLASS_NAME,"jobs-search__results-list").find_elements(By.TAG_NAME,'li') # return a list
print(len(jobs))
```

```python
#get jobs title, company, location, date
job_title = []
company_name = []
location = []
date = []
for job in jobs:
    job_title0 = job.find_element(By.CSS_SELECTOR,'h3').get_attribute('innerText')
    job_title.append(job_title0)

    company_name0 = job.find_element(By.CSS_SELECTOR,'h4').get_attribute('innerText')
    company_name.append(company_name0)

    location0 = job.find_element(By.CLASS_NAME,'job-search-card__location').get_attribute('innerText')
    location.append(location0)

    date0 = job.find_element(By.CSS_SELECTOR,'div>div>time').get_attribute('datetime')
    date.append(date0)
```

```python
#get jobs detail
job_desc = []
seniority = []
employment_type = []
job_function = []
industries = []
for item in range(len(jobs)):
    job_function0=[]
    industries0=[]

    # clicking job to view job details
    try:
        job_click_path = f'/html/body/div[1]/div/main/section[2]/ul/li[{item+1}]/a/div[1]/img'
        job_click = job.find_element(By.XPATH,job_click_path).click()
    except:
        job_click_path = f'/html/body/div[1]/div/main/section[2]/ul/li[{item+1}]/div/a'
    
    print(job_click_path)
    job_click = job.find_element(By.XPATH,job_click_path).click()
    time.sleep(7)
    
    #job description
    try:
        job_desc_path = '/html/body/div[1]/div/section/div[2]/div/section[1]/div/div/section'
        job_desc0 = job.find_element(By.XPATH,job_desc_path).get_attribute('innerText')
    except:
        job_desc_path = '/html/body/div[1]/div/section/div[2]/div/section[2]/div/div/section/div'
    
    job_desc0 = job.find_element(By.XPATH,job_desc_path).get_attribute('innerText')
    is_benefit = True
    try:
        assert 'Base pay range' in job_desc0
    except:
        is_benefit = False
        
    if is_benefit==True :
        job_desc_path = '/html/body/div[1]/div/section/div[2]/div/section[2]/div/div/section/div'
        job_desc0 = job.find_element(By.XPATH,job_desc_path).get_attribute('innerText')
        job_desc.append(job_desc0)
        job_desc_path2 = '/html/body/div[1]/div/section/div[2]/div/section[2]/div'
    else:
        job_desc.append(job_desc0)
        job_desc_path2 = '/html/body/div[1]/div/section/div[2]/div/section[1]/div'

    #seniority level
    try:
        seniority_path = job_desc_path2 + '/ul/li[1]/span'
        seniority_class_name = 'description__job-criteria-text description__job-criteria-text--criteria'
        seniority0 = job.find_element(By.XPATH,seniority_path).get_attribute('innerText')
        seniority.append(seniority0)
    except:
        seniority.append('') #handling if seniority is not available
    
    #Employment type
    try:
        employment_type_path = job_desc_path2 + '/ul/li[2]/span'
        employment_type_class_name = ''
        employment_type0 = job.find_element(By.XPATH,employment_type_path).get_attribute('innerText')
        employment_type.append(employment_type0)
    except:
        employment_type.append('') #handling if employment type is not available
    
    #job function
    try:
        job_function_path = job_desc_path2 + '/ul/li[3]/span'
        job_function_class_name = ''
        job_function0 = job.find_element(By.XPATH,job_function_path).get_attribute('innerText')
        job_function.append(job_function0)
    except:
        job_function.append('') #handling if job function is not available
    
    #industry
    try:
        industries_path = job_desc_path2 + '/ul/li[4]/span'
        industries_class_name = ''
        industries0 = job.find_element(By.XPATH,industries_path).get_attribute('innerText')
        industries.append(industries0)
    except:
        industries.append('')
```

```python
job_data = pd.DataFrame({'Date': date,
                        'Company': company_name,
                        'Title': job_title,
                        'Location': location,
                        'Description' : job_desc,
                        'Level': seniority,
                        'Type': employment_type,
                        'Function': job_function,
                        'Industry': industries
                        })
```
